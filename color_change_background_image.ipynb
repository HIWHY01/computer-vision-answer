# In this notebook you learn how to remove blue background 

```python
# Now it's time import our important libray need to remove blue screen

import matplotlib.pyplot as plt
import numpy as np 
import cv2  # this is a computer vision libray
```

```python
# Load image 
image = cv2.imread('/content/drive/MyDrive/dataset/pizza_bluescreen.jpg')
plt.imshow(image);
```

OOO know, this is not a blue screen. This is a Red screen. The question is why?
Becuse we are using OPENCV library and opencv color channel is BGR. So why our image display red. 

```python
# Now it's time to change color BGR to RGG.
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# See image 
plt.imshow(image);
```

```python
# let's see image infromation 
print ('This image type is -- ', type(image))
print ('Image Height, Width and Color Channel is ', image.shape)
```

**This image is ( 514 Height, 816 Width, 3 Color Channel)**

```python
# Define coor 
lower_blue = np.array([0, 0, 200])
upper_blue = np.array([250, 250, 255])
```

```python
# Define mask area 
mask = cv2.inRange(image, lower_blue, upper_blue)

# Now it's time see 
plt.imshow(mask, cmap='gray');
```

```python
# Mask the blue screen and only show pizza
masked_image = np.copy(image)
masked_image[mask != 0] = [0, 0, 0]

# Now it's time show image
plt.imshow(masked_image);
```
